```python
sns.countplot('Rating', data = df)
plt.title('Countplot by each Rate')
```

```python
df['Length'] = df['Review'].apply(len)
```

```python
sns.lineplot(data = df, x = 'Rating', y = 'Length')
```

```python
plt.figure(figsize = (12,7))
sns.scatterplot(data = df, x = 'Rating', y = 'Length', hue = 'Rating')
```
